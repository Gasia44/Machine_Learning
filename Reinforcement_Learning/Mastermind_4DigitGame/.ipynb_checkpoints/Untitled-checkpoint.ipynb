```python
def reply(secret, guess):
    """
    :param guess: a 4-digit array that oponent has guessed
    :return: reply how accurate the guess is with a tuple of 2 numbers
    """
    reply = np.zeros(2, dtype = np.int)
    sec = np.copy(secret)  
    count = 0
    for i in range(len(guess)):
        if guess[i] in sec:
            print(np.where(sec==guess[i])[0][0])
            sec[np.where(sec==guess[i])[0][0]] = -1
            count += 1
        print(sec)
            
    reply[0] = count

    reply[1] = ((guess == secret).sum())
    # TODO: Given a guess reply how accurate it was.
    return reply[0], reply[1]
```

```python
def asd(a,b,guess, pos_numbers):
    numbers = []
    for num in pos_numbers:
        if reply(num, guess) == (a,b): 
            numbers.append(num)
    return numbers
```
